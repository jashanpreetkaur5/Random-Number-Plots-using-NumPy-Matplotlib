```python
#-------------------------------------------------------------------------------
#C083872-Jashanpreet Kaur 
#-------------------------------------------------------------------------------

#importing libraries-----------------------------------------------------------               
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

#different sets of random numbers using NumPy----------------------------------
plot1 = np.random.normal(loc=0, scale=1, size=100)
plot2 = np.random.chisquare(df=3, size=100) 
plot3 = np.random.exponential(scale=1, size=100)
plot4 = np.random.normal(loc=2, scale=1, size=100)

# Set the plot style------------------------------------------------------------
style.use('fivethirtyeight')

# Create a figure with 5 subplots-----------------------------------------------
fig = plt.figure(figsize=(14, 9))

# Plot the first set of random numbers as a violin plot 
p1 = fig.add_subplot(2, 2, 1)
sns.violinplot(plot1, ax=p1, color='#5DA5DA', inner="stick", inner_linewidth=1.5, linewidth=2)
sns.stripplot(plot1, jitter=True, size=4, color='black', ax=p1)
p1.set_title('Violin Plot', fontsize=16, fontweight='bold', color='#5DA5DA')
p1.set_xlabel('Value', fontsize=12)
p1.set_ylabel('Density', fontsize=12)
p1.yaxis.grid(True, linestyle='--', alpha=0.7)
p1.spines['left'].set_linewidth(2)
p1.spines['bottom'].set_linewidth(2)

#-------------------------------------------------------------------------------
# Plot the second set of random numbers in the second subplot as a line plot
p2 = fig.add_subplot(2, 2, 2)
sns.lineplot(data=plot2, ax=p2, color='purple', linewidth=2.5,marker='o', markersize=6)

# Add a title and labels
p2.set_title('Line Plot', fontsize=16, fontweight='bold',color='purple')
p2.set_xlabel('Value', fontsize=12, fontweight='bold')
p2.set_ylabel('Density', fontsize=12, fontweight='bold')

# Set the tick labels color
p2.tick_params(axis='x', colors='black')
p2.tick_params(axis='y', colors='black')

# Set the grid lines color and style
p2.grid(axis='both', color='gray', linestyle='--')
#------------------------------------------------------------------------------

# Plot the third set of random numbers in the third subplot as a density plot-----
p3 = fig.add_subplot(2, 2, 3)
sns.kdeplot(plot3, color='#06d6a0', fill=True, ax=p3)
sns.rugplot(plot3, color='black', height=0.1, ax=p3)
p3.set_title('Density Plot', fontsize=16, fontweight='bold', color='#06d6a0')
p3.set_xlabel('Value', fontsize=12)
p3.set_ylabel('Density', fontsize=12)
p3.tick_params(axis='both', which='major', labelsize=10)
p3.yaxis.grid(True, linestyle='--', alpha=0.7)
p3.spines['left'].set_linewidth(2)
p3.spines['bottom'].set_linewidth(2)

#--------------------------------------------------------------------------------
# Add a 3D plot as the fourth subplot--------------------------------------------
p4 = fig.add_subplot(2, 2, 4, projection='3d')

# Generate some random data for the scatter plot--------------------------------
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
Z = np.random.normal(0, 1, 100)

# Plot the scatter plot
scatter = p4.scatter(X, Y, Z, c=np.arange(len(X)), cmap='rainbow', marker='o')
fig.colorbar(scatter, ax=p4)

# Add labels and title
p4.set_title('3D Scatter Plot', fontsize=12, fontweight='bold',color="red")
p4.set_xlabel('X')
p4.set_ylabel('Y')
p4.set_zlabel('Z')

#------------------------------------------------------------------------------

# Add a global title for the entire figure--------------------------------------
fig.suptitle('List of Random Number Distributions', fontsize=18, fontweight='bold', y=0.95)

# Adjust the spacing between subplots-------------------------------------------
plt.subplots_adjust(hspace=0.4, wspace=0.2)
plt.subplots_adjust(top=0.79)

# Save the plot as a PNG file---------------------------------------------------
plt.savefig('random_subplots.png', dpi=300, bbox_inches='tight')

# Display the plot in the notebook----------------------------------------------
plt.show()

#-------------------------------------------------------------------------------
```
